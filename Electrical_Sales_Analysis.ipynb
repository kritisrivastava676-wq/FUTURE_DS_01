# Business Sales Performance Analytics: Heavy Electricals
**Client:** Regional Electrical Goods & Heavy Power Systems
**Objective:** Transforming raw sales data into actionable business intelligence to answer core growth questions.

```python
import pandas as pd
import plotly.express as px

# 1. Read the raw data file
df = pd.read_csv('electrical_sales_data.csv')

# 2. Bulletproof Data Cleaning: Fix column names automatically
df.columns = df.columns.str.strip().str.replace(' ', '_').str.upper()

# 3. Process Dates and Calculate Revenue
df['DATE'] = pd.to_datetime(df['DATE'], dayfirst=True)
df['MONTH'] = df['DATE'].dt.strftime('%B %Y')
df['TOTAL_REVENUE'] = df['QUANTITY'] * df['UNIT_PRICE']

print("✅ Data cleaned and organized! Generating Business Dashboard...\n")

# 4. Chart 1: Top-Selling Products
p_data = df.groupby('PRODUCT_NAME')['TOTAL_REVENUE'].sum().reset_index().sort_values(by='TOTAL_REVENUE')
fig1 = px.bar(p_data, x='TOTAL_REVENUE', y='PRODUCT_NAME', orientation='h',
              title='<b>Top-Selling Products by Total Revenue</b>', color='TOTAL_REVENUE', color_continuous_scale='Blues')
fig1.show()

# 5. Chart 2: Revenue Trends Over Time
t_data = df.groupby('MONTH')['TOTAL_REVENUE'].sum().reset_index()
fig2 = px.line(t_data, x='MONTH', y='TOTAL_REVENUE', markers=True,
               title='<b>Business Revenue Trends Over Time</b>')
fig2.update_traces(line_color='#1b9e77', line_width=4, marker=dict(size=10))
fig2.show()

# 6. Chart 3: High-Value Category Breakdown
c_data = df.groupby('CATEGORY')['TOTAL_REVENUE'].sum().reset_index()
fig3 = px.pie(c_data, values='TOTAL_REVENUE', names='CATEGORY', hole=0.3,
              title='<b>High-Value Categories Profile Split</b>')
fig3.show()
```

## 📊 Executive Summary & Business Answers

**1. Which products generate the most revenue?**
While everyday items like 9W LED Bulbs have the highest transaction volume, the **1200Ah Industrial Battery Bank** and **900Ah Home Backup Systems** generate the vast majority of total financial revenue.

**2. How do sales change over time?**
Sales show massive revenue spikes specifically on days when heavy industrial equipment is sold. The timeline indicates that securing just one or two major contracts heavily skews the monthly profitability, highlighting a project-based revenue cycle rather than a steady retail cycle.

**3. Which categories are most profitable?**
The **Heavy Power** category is the absolute core economic engine of the business, far outperforming the combined revenue of Lighting, Wiring, and Switches.

**4. Where should the business focus to grow faster? (Actionable Recommendations)**
